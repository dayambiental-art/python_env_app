## Numpy

```python
# Importação do pacote Numpy
import numpy as np

# Geração de números aleatórios

# Valores máximos e mínimos
val_max, val_min = 10, 0.5

# Número de valores aleatórios
n = 1000

# Utilizando pacote Numpy e função uniform que está dentro da subpacote random
arr_aleatorio = np.random.uniform(val_min, val_max, n)

#print(arr_aleatorio)

#utilizando pacote numpy e função randint que está dentro do pacote random
arr_aleatorio_inteiro = np.random.randint(val_min, val_max, n)

#print (arr_aleatorio_inteiro)

```

```python
# Verificando o tamanho da matriz arr_aleatorio
# shape (formato da matriz)

print(arr_aleatorio.shape)
print(np.shape(arr_aleatorio))

#size (número de elementos)
print(arr_aleatorio.size)

# Utilizando o reshape para formatar o vetor em uma matriz 2 linhas e n colunas

mat = arr_aleatorio.reshape(2,-1)
print(np.shape(mat))
print(mat.size)
#print(mat)
```

```python
# Multiplicação de matrizes

#Multiplicação com arrays do mesmo tamanho
arr_resultado = arr_aleatorio_inteiro*arr_aleatorio

# Multiplicação com arrays de diferentes dimensões - ERRADO
multiplicador = np.random.randint(1,10,2)

arr_resultado2 = []
for multi in multiplicador:
    print(multi)
arr_resultado2 = arr_aleatorio_inteiro*multi
#print(arr_resultado2)

#arr_resultado3 = np.concatenate(arr_resultado2)
print(arr_resultado2.shape)

arr_resultado4 = np.hstack(arr_resultado2)
print(arr_resultado4.shape)

arr_resultado6 = arr_resultado4.transpose()
print(arr_resultado6.shape)

arr_resultado7 = np.random.rand(10,10,10)
print(arr_resultado7.shape)
```

```python
# Acessando valores da matriz
# Considerando que array_resultado7 é uma matriz de temperatura com as dimensões [lat,long,tempo]
# Determine a temperatura para o tempo igual a 1 em todas as latitudes e longitudes

arr_resultado7[:,:,0]

# Para todas as longitudes e primeira latitude e primeiro tempo
arr_resultado7[0,:,0]

# Todas as latitudes, primeira longitude e último tempo
arr_resultado7[:,0,-1]
arr_resultado7[:,0,np.shape(arr_resultado7)[2]-1]

# Para um ponto lat long no espaço e todo tempo
arr_resultado7[2,2,:]

# Uma região no espaço
arr_resultado7 [2:4,4:6,1]
```

```python
# Tamanho das dimensões da matriz

np.shape(arr_resultado7)[2]
```

```python
# Estatísticas com Numpy

min_absoluto = arr_resultado7.min()
print(min_absoluto)

# Mínimo para cada ponto lat long considerando todos os tempos
#axis considerando a dimensão de extração do dado (iniciando a partir de 0)
min_latlon = arr_resultado7.min(axis=2)
min_latlon = np.min(arr_resultado7,axis=2)
#print(min_latlon)

# Atribuindo um valor NaN (nada) dentro da matriz em algum lugar

arr_resultado7[2,3,1] = np.nan
#print(arr_resultado7)

# Minimo com NaN

print(np.nanmin(arr_resultado7))
```
