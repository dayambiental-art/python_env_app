## Tipos de variáveis

### Variável float - dupla precisão ou número real

```python
# variável float

b = 1.0
print (b)
print (type(b))
```

```python
# conversão de int para float
c = float(a)
print (c)
```

```python
# conversão de float para int
d = int(b)
print (d)
```

### Variável string - str - TEXTO

```python
# variável string
e = 'Júlia'
print (e)
print(type(e))
```

```python
# Converter um inteiro e um float para string
f = str(c)
g = str(d)
print('o valor de f é: '+ f +'e o valor de g: ' + g)
```

### Variável Booleana - Lógica

```python
# Comparação entre valores - usar: ==
h = (f == g)
print (h)
print (type(h))
```

```python
# Para comparar se as variáveis são diferentes de !=
i = (f!= g)
print (i)
print(type(i))
```

## LISTAS

```python
# Primeira lista

lista_a = ['Igor', 'Clarissa', 'Leonardo']
print (lista_a)
print (type(lista_a))
```

```python
# Criando uma lista de valores consecutivos

lista_consecutivos = list(range(0,100))
print (lista_consecutivos)
```

```python
# Primeiro valor
print(lista_consecutivos[0])

# Último valor
print(lista_consecutivos [99])
print(lista_consecutivos [-1])
```

```python
# Comprimento da lista
comprimento = len(lista_consecutivos)
print(comprimento)
```

```python
# Valos no meio da série
metade = comprimento/2
print(metade)
print(lista_consecutivos[int(metade)])
```

## Loop

```python
# usando range
for ii in range(0,10):
    print(ii)
```

```python
# usando uma lista e range
for ii in range (0,len(lista_a)):
    print(lista_a[ii])
```

```python
dic_2 = {'Igor': 10,
        'Marcos': 9,
        'Júlia': 5.75,
        'Clarissa': 5.75,
        'Leonardo': 5,
        'Leticia': 8}
for key,value in dic_2.items():
    if (value>5.75): 
    if (value>8):
        print(key + ' está aprovado com notão')
    elif value ==5.75:
        print(key + ' passou raspando')
    else:
        print (key + ' reprovado')
```

```python
# Loop em uma lista e acumulando valores especificos

lista_c = [100, 200, 10, 2, 1, 6, 22, 27]

#Acumulando valores maiores que 50
lista_maior_50 = []
for index, valor in enumerate(lista_c):
    if valor>50:
        print('índice da lista= ' +str(index))
        lista_maior_50.append(valor)

        print(lista_maior_50)
```

# Numpy

```python
# Usando lógica e Numpy

#Importando o numpy
import numpy as np
# Convertendo para Numpy
arr = np.array(lista_c)

print(arr)
print(type(arr))

# Selecionando valores maiores que 50 - retorna um array booleano
print(arr>50)

# imprimir apenas valores maiores que 50
print(arr[arr>50])
```

```python
for ii in arr:
    print(ii)
```

```python
# Média, mínimo, máximo, std, percentil, mediana
print('Média= '+ str(arr.mean()))
print('Mínimo= '+ str(arr.min()))
print('Máximo= '+ str(arr.max))
print ('Desvio= ' + str(arr.std))
print ('Percentil90= '+ str(np.percentile(arr,90)))
print ('Percentil50= ' + str(arr.percentile (arr, 50)))
print ('Mediana= '+ str(np. median(arr)))
```

```python
# Array de números aleatórios

val_min = 80
val_max = 600

array_temperatura_aleatoria = val_max-val_min*np.random.rand(1000) + val_min
print(array_temperatura_aleatoria)
```

```python
# Cinética

fx = 200*np.exp(array_temperatura_aleatoria) + array_temperatura_aleatoria**(-0.2)
print(fx)
```

## Figuras = Matplotlib

```python
#importando pacote
import matplotlib.pyplot as plt

fig, ax = plt.subplots(2)
ax[0].plot(fx)
ax[1].scatter(array_temperatura_aleatoria, fx, color= 'r')
ax[1].set_ylabel('fx')
```
