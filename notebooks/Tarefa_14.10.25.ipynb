```python
#Numpy

import numpy as np
```

```python
# Equação: u(rad).days(tot) = uW(rad).days(wet) + uDrad(mon).days(dry)

#radiação
valor_min= 10
valor_max= 80

u_rad_ = valor_max-valor_min*np.random.rand(100) + valor_min
print(u_rad_)
```

```python
#radiação solar em dias úmidos
uW = valor_max-valor_min*np.random.rand(100) + valor_min
print (uW)
```

```python
day1 = 1
day30 = 30

#dias totais

days_tot = day30-day1*np.random.rand(100) + day1
print (days_tot)
```

```python
#dias úmidos

days_wet = day30-day1*np.random.rand(100) + day1
print (days_wet)
```

```python
#Radiação Solar em dias úmidos

uD = valor_max-valor_min*np.random.rand(100) + valor_min
print (uD)
```

```python
#Dias secos 

days_dry = day30-day1*np.random.rand(100) + day1
print (days_dry)
```

```python
u_rad_ = (uW*days_wet + uD*days_dry)/days_tot
print (u_rad_)
```
